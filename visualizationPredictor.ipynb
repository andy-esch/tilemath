```python
%pylab inline
import pandas as pd
import numpy as np
import json
```

```python
earthquakes = pd.read_csv('data/all_day.csv', parse_dates=True)
earthquakes[:10]
```

```python
header = list(earthquakes.columns.values)
uniqueVals = {}
uniqueValCount = {}
maxVals = {}
minVals = {}
for h in header:
    uniqueVals[h] = set(earthquakes[h])
    uniqueValCount[h] = {"count": len(uniqueVals[h]), \
                         "type": type(earthquakes[h][0]).__name__, \
                         "diff":(len(earthquakes[h])-len(uniqueVals[h])), \
                         "max": max(uniqueVals[h]), \
                         "min": min(uniqueVals[h])
                         }
    if uniqueValCount[h]["type"] == "float64":
        uniqueValCount[h]["nanPercent"] = 100. * float(np.count_nonzero(np.isnan(earthquakes[h]))) / float(len(earthquakes[h]))

print json.dumps(uniqueValCount, indent=4, sort_keys=True)
```
